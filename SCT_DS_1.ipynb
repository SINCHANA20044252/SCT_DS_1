```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Age Group": ["0-20 years", "21-64 years", "65+ years"],
    "Population (Millions)": [512, 807, 98],
    "Percentage": [36.1, 57.0, 6.7]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8,6))
bars = plt.bar(df["Age Group"], df["Population (Millions)"], 
               color=["gold", "skyblue", "violet"], edgecolor="black")


for bar, value in zip(bars, df["Population (Millions)"]):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 10,   # 10 offset for clarity
             f"{value}M", ha='center', va='bottom', fontsize=10, fontweight="bold")


plt.title("India's Population Distribution by Age (2022)", fontsize=14, fontweight="bold")
plt.xlabel("Age Groups", fontsize=12)
plt.ylabel("Population (in Millions)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)


plt.show()
```

```python
plt.figure(figsize=(7,7))
colors = ["gold", "skyblue", "violet"]

plt.pie(df["Percentage"], 
        labels=df["Age Group"], 
        autopct="%1.1f%%", 
        startangle=90, 
        colors=colors, 
        wedgeprops={'edgecolor': 'black'})

plt.title("India's Population Share by Age (2022)", fontsize=14, fontweight="bold")
plt.show()
```
